The present project emphasizes the capacity of <b>word clouds</b> to function as a straightforward yet effective <i><b>instrument for obtaining valuable insights from extensive textual datasets</b></i>. By visually representing the <u>most frequently used words</u> in the constitution dataset, the word cloud facilitates the identification of key themes and concepts, thereby providing a deeper understanding of the text as a whole. As such, this project demonstrates the potential of word clouds as a <i>versatile</i> and <i>accessible analytical tool</i> for researchers and practitioners working with large textual datasets.
  
  
<br>


Information about the dataset you can find [here](https://www.kaggle.com/datasets/prasad22/constitution-dataset).

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import PIL.Image

# Read data
data = pd.read_csv('/kaggle/input/constitution-dataset/Constitution Dataset.csv') 

text = ' '.join(data['Preamble'].dropna().astype(str))

python_mask = np.array(PIL.Image.open("/kaggle/input/constitution-img/contitution.jpeg"))

wc = WordCloud(stopwords=STOPWORDS,
               mask=python_mask,
               background_color="black",
               min_font_size=7,
               max_words=400).generate(text)

plt.figure(figsize=(16, 10))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
